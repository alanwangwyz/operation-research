\begin{equation}
\left\{\begin{array}{l}{0.3 x_{1}+0.1 x_{2} \leq 2.7} \\ {0.5 x_{1}+0.5 x_{2}=6} \\ {0.6 x_{1}+0.4 x_{2} \geq 6}\end{array} x_{1} \geq 0 \text { and } x_{2} \geq 0\right.
\end{equation}

```python
from scipy import optimize
import numpy as np
c = np.array([-0.4,-0.5])
A = np.array([[0.3,0.1],[-0.6,-0.4]])
b = np.array([2.7,-6])
Aeq = np.array([[0.5,0.5]])
beq = np.array([6])
X0_bounds = [0,None]
X1_bounds = [0,None]

#求解
res = optimize.linprog(-c,A,b,Aeq,beq, bounds=(X0_bounds,X1_bounds))
print(res)
```
