```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
 
# Construct lines
x = np.linspace(0.0,10.0,11)
x1 = np.linspace(0.0,4.0,11)
y1 = 4.0
y2 = 6.0
y3 = (18-3*x)/2
y4 = 0.0
# Make plot
# plt.plot(y1, label=r'$y = 4.0$')
plt.hlines(y=6, xmin=0, xmax=6, colors = 'blue', label=r'$x_2= 6.0$')
plt.vlines(x=4, ymin=0, ymax=8, colors = 'green', label=r'$x_1 = 4.0$')
plt.plot(y3, label=r'$18=3x_1 + 2x_2$')
plt.xlim((0.0, 10.000))
plt.ylim((0.0, 10.000))
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
 
# Fill feasible region
y5 = np.minimum(y2, y3)
plt.fill_between(x, y4, y5, where=y3>2, color='grey', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
%config InlineBackend.figure_format = 'retina'
plt.show()
```

\begin{equation}
\left\{\begin{array}{c}{x_{1} \leq 4} \\ {2 x_{2} \leq 12} \\ {3 x_{1}+2 x_{2} \leq 18}\end{array} \quad x_{1} \geq 0 \text { and } x_{2} \geq 0\right.
\end{equation}

```python
from scipy import optimize
import numpy as np
c = np.array([3,5])
A = np.array([[1,0],[0,2]])
b = np.array([4,12])
Aeq = np.array([[3,2]])
beq = np.array([18])

#求解
res = optimize.linprog(-c,A,b,Aeq,beq)
print(res)
```
